#### Importing the required libraries

```python
import cv2
import matplotlib.pyplot as plt
```

#### Function to show image

```python
def show_img(img_name):
    plt.figure(figsize=(10, 10)) 
    plt.imshow(img_name)
    plt.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
    plt.show()
```

#### Reading the image

```python
image = cv2.imread('images/example_image.jpg')
show_img(image)
```

#### Converting the image to grayscale

```python
grayscale_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
show_img(grayscale_img)
```

#### Applying Canny edge detection on image

```python
image_edge_canny = cv2.Canny(grayscale_img, 50, 255)
```

```python
show_img(image_edge_canny)
```

#### Finding contours using CHAIN_APPROX_NONE method

```python
contours, hierarchy = cv2.findContours(image=image_edge_canny, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
```

#### Drawing contours on the image

```python
original_image_copy = image.copy()
cv2.drawContours(image=original_image_copy, contours=contours, contourIdx=-1, color=(0, 255, 0), 
                 thickness=2, lineType=cv2.LINE_AA);
```

```python
show_img(original_image_copy)
```

#### Finding contours using CHAIN_APPROX_SIMPLE method

```python
contours, hierarchy = cv2.findContours(image=image_edge_canny, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_SIMPLE)
```

#### Drawing contours

```python
original_image_copy = image.copy()
cv2.drawContours(image=original_image_copy, contours=contours, contourIdx=-1, color=(255, 0, 0), 
                 thickness=2, lineType=cv2.LINE_AA);
```

```python
show_img(original_image_copy)
```
